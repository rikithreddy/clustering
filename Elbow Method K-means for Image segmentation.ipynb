# Elbow method K-means
This notebook is for illustrating elbow method to determine number of clusters for kmeans. The idea is to use an image with specific number of colors. So when we cluster on the image we know how many clusters to expect. 

## Import all required libraries
* cv2 for loading images
* numpy for martix manuplation
* Kmeans algorithm from sklearn
* matplotlib for plotting

```python
import cv2
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

## Load the image

Notice that the image has only 5 colors ie red, blue, yellow, black, white. So the best value of k should be 5 when we cluster the image.


```python
img = cv2.imread("data/Primary Colors.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## Reshape the image for clustering

```python
x, y, z = img.shape
img = img.reshape(x*y, z)
img.shape
```

## Run kmeans varying the number of clusters from 1 to 20

For each value of k also capture the kmeans inertia, which is the mean squared error of the points from their respective centers obtained by clustering

```python
cluster_mss = []
k_values = range(1, 20)
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans = kmeans.fit(img)
    cluster_mss.append(kmeans.inertia_)
```

## Plot the MSS

From the plot below we can see that the MSS value has little or no change after k=5 (which is like an elbow). Hence we choose the number of clusters = 5. This is actually true as there were only 5 colors in the original image.

```python
plt.plot(k_values, cluster_mss, "ro")
plt.plot(k_values, cluster_mss, "r")
plt.title("Elbow curve")
plt.ylabel("Mean Squared error for each point wrt center")
plt.ylabel("Value of K")
```
